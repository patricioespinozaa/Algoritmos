```python
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
# Configuración estética para gráficos
sns.set(style="whitegrid")
data = pd.read_csv('experimento1_resultados.csv',)
exp_2 = pd.read_csv('experimento2_resultados.csv')
exp_3 =  pd.read_csv('experimento3_resultados.csv')
exp_4 =  pd.read_csv('experimento4_resultados.csv')
pd.options.display.float_format = '{:.6f}'.format
#data.head()
#exp_2.head()
#exp_3.head()
exp_4.head()

```

Visualización de Tiempos de Inserción visualizacion tiempos de insercion Costo Promedio de Búsqueda

```python
fig, axes = plt.subplots(1, 3, figsize=(18, 6), sharex=True)

# Tiempo de Inserción
axes[0].plot(data['N'], data['Tiempo_Insercion_ABB'], label='ABB', marker='o', color='b')
axes[0].plot(data['N'], data['Tiempo_Insercion_Splay'], label='Splay', marker='o', color='r')
axes[0].set_title("Tiempo de Inserción")
axes[0].set_xlabel('N (Número de elementos)')
axes[0].set_ylabel('Tiempo (ms)')
axes[0].legend()
axes[0].grid(True)

# Tiempo de Búsqueda
axes[1].plot(data['N'], data['Tiempo_Busqueda_ABB'], label='ABB', marker='o', color='b')
axes[1].plot(data['N'], data['Tiempo_Busqueda_Splay'], label='Splay', marker='o', color='r')
axes[1].set_title("Tiempo de Búsqueda")
axes[1].set_xlabel('N (Número de elementos)')
axes[1].legend()
axes[1].grid(True)

# Costo Promedio de Búsqueda
axes[2].plot(data['N'], data['Costo_Promedio_ABB'], label='ABB', marker='o', color='b')
axes[2].plot(data['N'], data['Costo_Promedio_Splay'], label='Splay', marker='o', color='r')
axes[2].set_title("Costo Promedio de Búsqueda")
axes[2].set_xlabel('N (Número de elementos)')
axes[2].legend()
axes[2].grid(True)

plt.suptitle("Experimento 1: Comparación de Desempeño entre ABB y Splay")
plt.show()
```
